## Poisson

```python
from scipy.stats import poisson
```

```python
np.random.seed(42)
k=np.arange(0,21)
Data=np.random.poisson(.61,size=200)
Data
```

```python
sns.histplot(x=Data)
```

```python
poisson.pmf(k,.61)
```

```python
probabilities = [np.mean(Data == i) for i in range(21)]
for i, prob in enumerate(probabilities):
    print(f"Probability of {i} soldiers per year: {prob:.4f}")
```

```python
print(poisson.stats(200,.61))
sns.countplot(x=Data)
```

## Normal (Gaussian) Distribution


```python
from scipy.stats import norm
from scipy.stats import zscore
mu=527
std=112
```

```python
x=np.arange(0,1001)
cdf = norm.cdf(x, loc=mu, scale=std)
x=np.arange(0,1001)
pdf = norm.pdf(x, loc=mu, scale=std)
sns.lineplot(x=x,y=cdf)
plt.show()
sns.lineplot(x=x,y=pdf)
plt.show()
```

```python
np.random.seed(101)
D=np.random.normal(mu,std,size=1000)
mean=np.mean(D)
mean
print(mean)
var=np.var(D)
print(var)
sns.histplot(x=D,kde=True)
plt.show()
std=np.std(D)
print(std)
```

```python
below=norm.cdf(500,loc=mu,scale=std)
print(below)
print(1-below)
```
