```python
# Import Libraries
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Read employees.csv file.
df = pd.read_csv("data/employees.csv")
df.head()
```

```python
# Display column counts.
print(f"Number Of Columns = {len(df.columns)}\n")
print("\t\tInfo\t\t")
df.info()
print("\n\tCount Value In Each Columns\t\t")
df.count()
```

```python
# Remove any row contains null and display column counts
df.dropna(inplace = True)
df.isnull().sum()
```

## Simple Random Sample (SRS)


```python
# Select simple random sample of size=50.
srs_sample = df.sample(n=50, random_state=42)
# Select 50% SRS.
srs_sample_ = df.sample(frac=0.5, random_state=42)
```

```python
# Display column counts in both cases.
print("\nSample One:\n")
print(srs_sample.count())
print("\nSample Two:\n")
print(srs_sample_.count())
```

## Systematic Sampling


```python
#Select a systematic sample with 25 interval
sample_25 = df[::25]
sample_25
```

```python
#Select a systematic sample with 50 interval
sample_50 = df[::50]
sample_50
```

```python
#Select a systematic sample with 75 interval
sample_75 = df[::75]
sample_75
```

```python
#Select a systematic sample with 100 interval
sample_100 = df[::100]
sample_100
```

## Stratified Sampling

```python
#Perform a countplot and pieplot for Gender column.
sns.countplot(data = df, x='Gender');
plt.show()
plt.pie(df['Gender'].value_counts().values,labels=df['Gender'].value_counts().index,autopct='%1.1f%%');
```

```python
# Select 20% startified sample from data based on Gender.
sample_20_ = df.groupby('Gender').apply(lambda x: x.sample(frac = 0.2))
sample_20_
```

```python
# Perform a countplot and pieplot for Gender column for the data sample.
sns.countplot(data = sample_20_, x='Gender');
plt.show()
plt.pie(sample_20_['Gender'].value_counts().values,labels=sample_20_['Gender'].value_counts().index,autopct='%1.1f%%');
```
