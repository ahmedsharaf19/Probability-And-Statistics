```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy.stats import norm
from scipy.stats import uniform
%matplotlib inline
```

```python
# Take a 10000 samples of size 1 from a continuous uniform distribution a=1, b=11 
# Plot the sampling distribution of the sample mean.
sample_one_means = []
for i in range(10000):
    s = np.random.uniform(low=1, high=12, size=1)
    sample_one_means.append(s.mean())
print(f"Mean Of Sample Distribution : {np.mean(sample_one_means)}")
print(f"Var Of Sample Distribution : {np.var(sample_one_means)}")
print(f"statistic Of Population  : {uniform.stats(1,10)}")
```

```python
sns.histplot(data = sample_one_means, kde = True)
```

```python
# Take a 10000 samples of size 100 from a continuous exponential distribution a=1, b=11 
# Plot the sampling distribution of the sample mean.
from scipy.stats import expon
np.random.seed(42)
sample_means = []
for i in range(10000):
    exp = np.random.exponential(scale = 10000, size = 100)
    sample_means.append(exp.mean())
print(f"Mean Of Sample Distribution : {np.mean(sample_means)}")
print(f"Var Of Sample Distribution : {np.var(sample_means)}")
print(f"statistic Of Population  : {expon.stats(10000)}")
sns.histplot(data = sample_means, kde = True);
```
